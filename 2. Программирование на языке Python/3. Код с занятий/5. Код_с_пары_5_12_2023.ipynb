## Код программы

```python
import PySimpleGUI as sg

sg.theme("SystemDefault1")


def gms_to_deg(g: int = None, m: int = None, s: float = None) -> float:
    '''
    Функция принимает на вход градусы, минуты, секунды и возвращает десятичные градусы (deg)
    :param g: Градусы в целочисленном представлении
    :param : Минуты в целочисленном представлении
    :param deg: Секунды в вещественном представлении
    :return: Десятичные градусы
    '''

    # В этом коде используется конструкция try-except для обработки возможных
    # исключений при выполнении кода если внутри блока try произойдет ошибка,
    # то выполняется код блока except
    try:
        # расчет десятичных градусов
        # переменной deg присваивается результат вычисления
        deg: float = int(g) + int(m) / 60 + float(s) / 3600
    except:
        # если произошло исключение (ошибка), то возвращается значение None
        return None

    # если ошибок не произошло, то возвращается значение deg
    return deg


def deg_to_gms(deg=None):
    try:
        deg = float(deg)
        g = int(deg)
        m = int((deg - g) * 60)
        s = (((float(deg) - g) * 60) - m) * 60
        return g, m, round(s, 5)
    except:
        return None


# Определение функции `menu_window()` - главное окно программы
def menu_window():
    # Создание переменной `layout` с описанием расположения элементов интерфейса окна
    layout = [
        # Создание кнопки "GMS to Deg" с ключом `1` и расширяемость по оси `y` и `x`
        [sg.Button("GMS to Deg", key=1, expand_y=True, expand_x=True)],
        # Создание кнопки "Deg to GMS" с ключом `2` и расширяемость по оси `y` и `x`
        [sg.Button("Deg to GMS", key=2, expand_y=True, expand_x=True)],
        # Создание кнопки "Direct Task" с ключом `3` и расширяемость по оси `y` и `x`
        [sg.Button("Direct Task", key=3, expand_y=True, expand_x=True)],
        # Создание кнопки "Reverse Task" с ключом `4` и расширяемость по оси `y` и `x`
        [sg.Button("Reverse Task", key=4, expand_y=True, expand_x=True)],
    ]

    # Возвращение окна с заголовком 'GeoCalc', указанным расположением элементов `layout`,
    # размером (250, 200), возможностью изменять размер окна и окончательным формированием интерфейса окна
    return sg.Window('GeoCalc', layout, size=(250, 200), resizable=True, finalize=True)


# Определяем функцию gms_window()
def gms_window():
    # Задаем переменную layout, которая содержит в себе описание интерфейса программы
    layout = [
        # Создаем первую строку интерфейса, содержащую текст "Введите градусы:" и поле для ввода градусов
        [sg.Text("Введите градусы: ", expand_y=True, expand_x=True), sg.InputText(key='g', enable_events=True)],
        # Создаем вторую строку интерфейса, содержащую текст "Введите минуты:" и поле для ввода минут
        [sg.Text("Введите минуты: ", expand_y=True, expand_x=True), sg.InputText(key='m', enable_events=True)],
        # Создаем третью строку интерфейса, содержащую текст "Введите секунды:" и поле для ввода секунд
        [sg.Text("Введите секунды: ", expand_y=True, expand_x=True), sg.InputText(key='s', enable_events=True)],
        # Создаем четвертую строку интерфейса, содержащую текст "Десятичные градусы:" и поле для вывода результата
        [sg.Text("Десятичные градусы: ", expand_y=True, expand_x=True),
         sg.Text("", key='answer', expand_y=True, expand_x=True, enable_events=True)],
    ]

    # Возвращаем окно с названием "GMS to Deg", содержащее описанный выше интерфейс
    return sg.Window('GMS to Deg', layout, size=(250, 200), resizable=True, finalize=True)


# Определяем функцию gms_window()
def deg_window():
    # Создаем список "layout", который будет описывать расположение элементов на интерфейсе
    layout = [
        # Создаем первую строку интерфейса, содержащую текст "ВВведите десятичные градусы: "
        [sg.Text("Введите десятичные градусы: ", font='Roboto 10 bold', justification='center', expand_y=True,
                 expand_x=True)],
        # Добавляем текстовое поле на интерфейс, в котором пользователь сможет вводить десятичные градусы
        [sg.InputText(key='deg', enable_events=True, expand_x=True)],
        # Добавляем текстовый элемент на интерфейс с выводом градусов, минут и секунд
        [sg.Text('Градусы, минуты, секунды: ', font='Roboto 10 bold', justification='center', expand_y=True,
                 expand_x=True)],
        # Добавляем пустой текстовый элемент на интерфейс, в котором будет отображаться результат преобразования
        [sg.Text('', key='GMS', expand_y=True, expand_x=True, font='Roboto 10 bold', justification='center',
                 enable_events=True)],
        # Добавляем две кнопки "Clear" и "Copy" на интерфейс
        [sg.Button('Clear', expand_y=True, expand_x=True), sg.Button('Copy', expand_y=True, expand_x=True)]
    ]

    # Возвращаем окно интерфейса с заданными параметрами
    return sg.Window('Deg to GMS', layout, size=(250, 200), resizable=True, finalize=True)


# Создаем три переменные window1, window2 и window3 и присваиваем им значения, возвращаемые функцией menu_window().
# menu_window() - это функция, которая создает окно с главным меню.
window1, window2, window3 = menu_window(), None, None


while True:
    window, event, values = sg.read_all_windows()
    # Считываем окно, на котором произошло событие, само событие и значения элементов интерфейса.
    # sg.read_all_windows() возвращает кортеж из трех элементов: окно, событие, значения.

    print(window, event, values)
    # Выводим на печать окно, событие и значения для отладки.

    if window == window1:
        # Если текущее окно - window1 (главное меню).

        if event == sg.WINDOW_CLOSED:
            # Если произошло событие закрытия окна.
            break
            # Выход из цикла while True.

        if event == 1:
            # Если произошло событие с идентификатором 1.

            window2 = gms_window()
            # Создаем окно с помощью функции gms_window() и присваиваем его переменной window2.
            # gms_window() - это функция, которая создает окно для ввода градусов, минут и секунд.

        if event == 2:
            # Если произошло событие с идентификатором 2.

            window3 = deg_window()
            # Создаем окно с помощью функции deg_window() и присваиваем его переменной window3.
            # deg_window() - это функция, которая создает окно для ввода градусов.

    if window == window2:
        # Если текущее окно - window2 (окно для ввода градусов, минут и секунд).
        if event == sg.WINDOW_CLOSED:
            # Если произошло событие закрытия окна window2.
            window2.close()
            # Закрываем окно window2.
            window2 = None
            # Присваиваем переменной window2 значение None.

        if event in ['g', 'm', 's']:
            # Если произошло одно из событий 'g', 'm', 's'.

            answer = gms_to_deg(values['g'], values['m'], values['s'])
            # Создаем переменную answer и присваиваем ей результат работы функции gms_to_deg()
            # с передачей в нее значений из элементов интерфейса 'g', 'm', 's'.
            # gms_to_deg() - это функция, которая конвертирует градусы, минуты и секунды в десятичные градусы.

            if answer is not None:
                # Если ответ не равен None (произошла успешная конвертация).
                window2['answer'].update(str(answer))
                # Обновляем значение элемента интерфейса 'answer' в окне window2.

        if event == 'answer':
            # Если произошло событие с элементом интерфейса 'answer'.

            sg.clipboard_set(window2['answer'].get())
            # Копируем значение элемента интерфейса 'answer' в буфер обмена.

            sg.popup_no_buttons('Данные успешно скопированы!', auto_close=True, auto_close_duration=1)
            # Выводим всплывающее сообщение без кнопок с текстом "Данные успешно скопированы!".
            # auto_close=True означает, что сообщение будет автоматически закрыто через указанное время
            for  i in ['g', 'm', 's', 'answer']:
                window2[i].update('')
            #  Очищаем поля для ввода данных

    # Этот блок кода проверяет, является ли окно, с которым взаимодействует пользователь, window3.
    # Если это окно, то выполняются определенные действия.
    if window == window3:
        # Если пользователь закрывает окно, то оно закрывается и переменной window3 присваивается значение None.
        if event == sg.WINDOW_CLOSED:
            window3.close()
            window3 = None
        answer = deg_to_gms(values['deg'])
        # Вызывается функция deg_to_gms, которая преобразует значение в градусах в градусы, минуты и секунды.
        # Если событие было вызвано элементом с именем 'deg'.
        if event == 'deg':
            # Если результат преобразования не является None.
            if answer is not None:
                # Форматирует результат преобразования в виде 'градусы° минуты′ секунды″'.
                answer = f'{answer[0]}° {answer[1]}′ {answer[2]}″'
                # Обновляет значение элемента 'GMS' в окне window3, отображая преобразованное значение.
                window3['GMS'].update(answer)

        # Если событие было вызвано элементом с именем 'Clear'.
        if event == 'Clear':
            # Очищает поля элементов 'deg' и 'GMS' в окне window3.
            for i in ['deg', 'GMS']:
                window3[i].update('')

        # Если событие было вызвано элементом с именем 'Copy' или 'GMS' и результат преобразования не является None.
        if event in ['Copy', 'GMS'] and answer is not None:
            # Копирует результат преобразования в буфер обмена.
            sg.clipboard_set(answer)
            # Показывает всплывающее окно с сообщением о успешном копировании данных.
            sg.popup_no_buttons('Данные успешно скопированы!', auto_close=True, auto_close_duration=1)

window1.close() # Закрывает окно window1 сразу после выхода из цикла
```
